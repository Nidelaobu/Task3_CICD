## 4. Cleaning & Transformation (2011)

### **Feature Engineering Decisions:**

#### **1. Date Extraction:**
- **Extract:** `day_of_year`, `month`, `weekday` from `dteday`
- **Cyclical Encoding:** Convert month/weekday to `sin/cos` components
  - **Why?** December (month=12) and January (month=1) are close in time but far numerically
  - **How?** `month_sin = sin(2Ï€ Ã— month/12)`, `month_cos = cos(2Ï€ Ã— month/12)`
  - **Benefit:** Model learns "December â†’ January" cyclical pattern

#### **2. Handling `atemp` (Feels-Like Temperature):**
- **Problem:** `temp` and `atemp` have **correlation > 0.99** (multicollinearity)
- **Impact:** In linear regression, unstable coefficients (high variance)
- **Solution:** Drop `atemp`, keep `temp` (slightly more direct physical measurement)

#### **3. Power Transformation (`hum` â†’ `hum_transformed`):**
- **Purpose:** Fix skewed distributions (make them Gaussian)
- **Method:** Yeo-Johnson Power Transform (works with 0 values)
- **Why?** Linear regression assumes normally distributed residuals
- **When to use:** If histogram shows heavy skew (long tail)

#### **4. Log Transformation (`windspeed` â†’ `windspeed_log`):**
- **Purpose:** Compress outliers, stabilize variance
- **Formula:** `log(windspeed + Îµ)` where Îµ=0.001 (avoid log(0))
- **Why?** Wind speed has extreme values that can dominate the model

```python
# Extract temporal features from dteday
df_2011['day_of_year'] = pd.to_datetime(df_2011['dteday'], dayfirst=True).dt.dayofyear
df_2011 = df_2011.drop('dteday', axis=1)
# month captures seasonal patterns; day_of_year 
# may help model learn annual trends


# Cyclical encoding for month
df_2011['month_sin'] = np.sin(2 * np.pi * df_2011['mnth'] / 12)
df_2011['month_cos'] = np.cos(2 * np.pi * df_2011['mnth'] / 12)
df_2011 = df_2011.drop('mnth', axis=1)

# Cyclical encoding for weekday
df_2011['weekday_sin'] = np.sin(2 * np.pi * df_2011['weekday'] / 7)
df_2011['weekday_cos'] = np.cos(2 * np.pi * df_2011['weekday'] / 7)
df_2011 = df_2011.drop('weekday', axis=1)
# Weekend demand pattern is different from weekdays
# Linear models will treat weekday=6 as "far" from weekday=0, but they're adjacent days


print(f"Total samples: {len(df_2011)}")
print(f"Missing values:\n{df_2011.isnull().sum()}") 
```

### Outlier Analysis

**Why Check for Outliers:**  
Weather sensors can malfunction or record extreme events. Outliers can:
- Skew model coefficients (especially in linear models)
- Reduce model generalization

**Our Approach:**  
We'll visualize distributions using boxplots to identify extreme values. However, we WON'T remove outliers blindly because:
1. Extreme weather (e.g., snowstorms) legitimately affects demand
2. Tree-based models (Random Forest) are robust to outliers
3. Removing data reduces our already-small dataset (365 days)

If outliers exist, we'll note them but proceed with full data.

### Train-Test Split Strategy

**CRITICAL: Temporal Split (Not Random)**

We use an 80/20 temporal split (first 292 days = train, last 73 days = test) because:

1. **Preserves Time Ordering:** Random splitting would leak future information into training (data leakage), artificially inflating performance.

2. **Simulates Real Deployment:** In production, we train on past data and predict future demand. This split mimics that scenario.

3. **Strictest Test:** The model has NEVER seen late-2011 patterns (Nov-Dec), making our RÂ²=0.80 result more trustworthy.

**Trade-off:** Test set is small (73 days), so metrics have higher variance. A k-fold cross-validation would give more stable estimates, but we prioritize time-aware evaluation.

```python
# ---  TRAIN-TEST SPLIT ---
# Use temporal split (not random!) for time series data
split_idx = int(0.8 * len(df_2011))
train_df = df_2011.iloc[:split_idx].copy()
test_df = df_2011.iloc[split_idx:].copy()

X_train = train_df.drop('cnt', axis=1)
y_train = train_df['cnt']
X_test = test_df.drop('cnt', axis=1)
y_test = test_df['cnt']
print(f"Train Shape: X={X_train.shape}, y={y_train.shape}")
print(f"Test Shape:  X={X_test.shape},  y={y_test.shape}")
```

```python
def cap_outliers(df, col, method='iqr', threshold=1.5):
    """
    Caps outliers using IQR or Z-Score method.
    
    Parameters:
    - df: DataFrame
    - col: Column name (str)
    - method: 'iqr' or 'z-score'
    - threshold: Multiplier for IQR (usually 1.5) or Std Dev for Z-score (usually 3)
    """
    df_clean = df.copy()
    
    if method == 'iqr':
        # Calculate IQR
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Define bounds
        lower_bound = Q1 - (threshold * IQR)
        upper_bound = Q3 + (threshold * IQR)
        
        # Cap values (Winsorization)
        df_clean[col] = np.where(df_clean[col] < lower_bound, lower_bound,
                        np.where(df_clean[col] > upper_bound, upper_bound, df_clean[col]))
        
    elif method == 'z-score':
        # Calculate Z-score bounds
        mean = df_clean[col].mean()
        std = df_clean[col].std()
        
        lower_bound = mean - (threshold * std)
        upper_bound = mean + (threshold * std)
        
        # Cap values
        df_clean[col] = np.where(df_clean[col] < lower_bound, lower_bound,
                        np.where(df_clean[col] > upper_bound, upper_bound, df_clean[col]))
    
    return df_clean

# --- APPLYING THE RECOMMENDATIONS ---

# 2. Apply IQR to Windspeed & Humidity (Skewed data)
X_train = cap_outliers(X_train, 'windspeed', method='iqr')
X_test = cap_outliers(X_test, 'windspeed', method='iqr')
X_train = cap_outliers(X_train, 'hum', method='iqr')
X_test = cap_outliers(X_test, 'hum', method='iqr')
                       
# 3. Apply Z-Score to Temperature (Normal-ish data)
X_train = cap_outliers(X_train, 'temp', method='z-score', threshold=3)
X_test = cap_outliers(X_test, 'temp', method='z-score', threshold=3)
```

```python
non_binary_cols = ['temp', 'hum','windspeed']


# 1. Melt the specific columns into long format
df_melted = X_train.melt(value_vars=non_binary_cols, 
                    var_name='Variable', 
                    value_name='Value')

# 2. Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x='Variable', y='Value')
plt.title('Distribution of Weather Variables')
plt.show()
```

```python
# --- 1. Log Transform Windspeed (Right Skewed) ---
X_train['windspeed_log'] = np.log1p(X_train['windspeed'])
X_test['windspeed_log'] = np.log1p(X_test['windspeed'])


# --- 2. Yeo Johnson humidity (left Skewed) ---
# Reshape data to 2D array for sklearn
yj = PowerTransformer(method='yeo-johnson')

yj.fit(X_train[['hum']])
X_train['hum_transformed'] = yj.transform(X_train[['hum']])
X_test['hum_transformed'] = yj.transform(X_test[['hum']])
# --- Visualization ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot Windspeed
sns.histplot(X_train['windspeed_log'], kde=True, ax=axes[0], color='green')
axes[0].set_title('Windspeed: Log Transformed')

# Plot Humidity
sns.histplot(X_train['hum_transformed'], kde=True, ax=axes[1], color='purple')
axes[1].set_title(f'Humidity: Yeo-Johnson Transformed\n(Skew: {X_train["hum_transformed"].skew():.2f})')

plt.tight_layout()
plt.show()
```

```python
# 1. Initialize
scaler = StandardScaler()

# 2. Define columns
cols_to_scale = ['temp', 'hum_transformed', 'windspeed_log']
scaled_col_names = [f"{col}_scaled" for col in cols_to_scale]

# 3. Fit and Transform
X_train_scaled_raw = scaler.fit_transform(X_train[cols_to_scale])
X_test_scaled_raw = scaler.transform(X_test[cols_to_scale])

# 4. Convert back to DataFrame (Crucial for concat)
X_train_scaled_df = pd.DataFrame(X_train_scaled_raw, columns=scaled_col_names, index=X_train.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled_raw, columns=scaled_col_names, index=X_test.index)

# 5. Concatenate
X_train = pd.concat([X_train, X_train_scaled_df], axis=1)
X_test = pd.concat([X_test, X_test_scaled_df], axis=1)

# 6. --- CHECK THE RESULT ---
print("Means (should be ~0):\n", X_train[scaled_col_names].mean().round(2))
print("\nStd Devs (should be 1):\n", X_train[scaled_col_names].std().round(2))
```

## Residual Plot Analysis

The residual plot is the primary tool for determining if model errors are random (**homoscedastic**) or follow a systematic pattern (**heteroscedastic**).

### Error Calculation
The residuals are calculated as the difference between observed and predicted values:
$$Residual = y_{test} - y_{pred\_test}$$

### Observation Summary
* **Central Tendency:** The points are scattered around the horizontal dashed line at **0**, representing the baseline for perfect predictions.
* **Error Distribution:**
    * **Low Volume (0 - 3000):** Residuals are relatively tightly clustered, indicating higher model stability.
    * **High Volume (> 3000):** The spread of residuals widens significantly as predicted values increase.
* **Outliers:** Notable errors include one significant **under-prediction** (residual near -2500) and several **over-predictions** exceeding +1000.

### 5.2 Ridge & RandomForest

```python
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor

# --- Experiment 2: Ridge Regression (Regularization) ---
# Why: Penalizes large coefficients to reduce overfitting from correlated features
ridge_model = Ridge(alpha=1.0) # Alpha is the strength of regularization
ridge_model.fit(X_train[available_features], y_train)
ridge_pred = ridge_model.predict(X_test[available_features])

# --- Experiment 3: Random Forest (Non-Linear) ---
# Why: Captures complex patterns (e.g., rentals drop if it's too hot OR too cold)
# Constraint: I limit max_depth=5 to prevent overfitting on this small dataset
rf_model = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf_model.fit(X_train[available_features], y_train)
rf_pred = rf_model.predict(X_test[available_features])

print("Challenger Models Trained!")
```

### Experiment 2: Regularized Regression (Ridge)

Why it is performed: Our feature set includes highly correlated variables (Multicollinearity), such as temp and atemp, or season and month.

Improvement Expected: Ridge Regression adds an L2 Penalty to the loss function. We expect this to shrink the coefficients of redundant features, reducing the model's Variance and making it more stable on the unseen Test set.

### Experiment 3: Non-Linear Model (Random Forest Regressor)

Why it is performed: Bike demand follows non-linear patterns (e.g., "Demand is high ONLY IF it is sunny AND not a holiday"). Linear models cannot easily capture these "If/Then" thresholds.

Improvement Expected:

Pros: Significant reduction in Bias. The model can learn complex, jagged decision boundaries.

Trade-off: High Variance (Risk of Overfitting). Given the small sample size of 2011 (~365 days), a tree model might memorize "noise" rather than the trend. We limited max_depth=5 to control this.

```python
# Helper function to calculate metrics
def get_metrics(name, y_true, y_pred):
    return {
        'Model': name,
        'RMSE': np.sqrt(mean_squared_error(y_true, y_pred)),
        'R2': r2_score(y_true, y_pred),
        'MAE': mean_absolute_error(y_true, y_pred)
    }

# Gather results
results = [
    get_metrics('Baseline (Linear)', y_test, y_pred_test),
    get_metrics('Ridge Regression', y_test, ridge_pred),
    get_metrics('Random Forest', y_test, rf_pred)
]

# Create DataFrame
results_df = pd.DataFrame(results)

# Display sorted by best Test R2
print("\n--- Model Leaderboard (Task 1: 2011 Data) ---")
display(results_df.sort_values(by='R2', ascending=False))
```

**Decision:** Random Forest selected as the champion model for deployment.

## 7. Logging all of the models and artificats into MLFlow (2011 Dataset)

```python
import mlflow
import mlflow.sklearn
from mlflow.models.signature import infer_signature
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# 1. Setup MLflow with Database
mlflow.set_tracking_uri("sqlite:///mlflow.db")
mlflow.set_experiment("Task1_BikeSharing_2011")

# 2. Define Pipelines
pipe_linear = Pipeline([('scaler', StandardScaler()), ('model', LinearRegression())])
pipe_ridge = Pipeline([('scaler', StandardScaler()), ('model', Ridge(alpha=1.0))])
pipe_rf = Pipeline([('scaler', StandardScaler()), ('model', RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42))])

models_to_log = [
    ("Baseline_Linear", pipe_linear, {}),
    ("Ridge_Regression", pipe_ridge, {"alpha": 1.0}),
    ("Random_Forest", pipe_rf, {"n_estimators": 100, "max_depth": 5})
]

print("--- Starting MLflow Logging ---")

best_r2 = -1
best_run_id = None
best_model_name = ""
best_model_uri = ""

# Features list (Must match X_train columns exactly)
features_for_pipeline = ['temp', 'hum_transformed', 'windspeed_log', 'day_of_year', 
                         'season', 'holiday', 'workingday', 'weathersit', 'month_sin', 'month_cos', 'weekday_sin', 'weekday_cos']

for name, pipeline, params in models_to_log:
    RUN_ID = None
    with mlflow.start_run(run_name=name) as run:
        RUN_ID = run.info.run_id
        # A. Fit
        pipeline.fit(X_train[features_for_pipeline], y_train)
        
        # B. Predict
        pred_test = pipeline.predict(X_test[features_for_pipeline])
        pred_train = pipeline.predict(X_train[features_for_pipeline])
        
        # C. Calculate Metrics (Added MAE here)
        r2_test = r2_score(y_test, pred_test)
        rmse_test = np.sqrt(mean_squared_error(y_test, pred_test))
        mae_test = mean_absolute_error(y_test, pred_test)
        r2_train = r2_score(y_train, pred_train)
        rmse_train = np.sqrt(mean_squared_error(y_train, pred_train))
        mae_train = mean_absolute_error(y_train, pred_train)  
        
        # D. Log Params & Metrics
        mlflow.log_params(params)
        mlflow.log_metric("test_rmse", rmse_test)
        mlflow.log_metric("test_r2", r2_test)
        mlflow.log_metric("test_mae", mae_test)
        mlflow.log_metric("train_rmse", rmse_train)
        mlflow.log_metric("train_r2", r2_train)
        mlflow.log_metric("train_mae", mae_train) 
        
        # E. Log Model
        signature = infer_signature(X_test[features_for_pipeline], pred_test)
        
       
        model_info = mlflow.sklearn.log_model(
            sk_model=pipeline, 
            artifact_path="model",
            signature=signature,
            input_example=X_train[features_for_pipeline].head()
        )
        
        print(f"Logged {name}: R2={r2_test:.4f}, MAE={mae_test:.2f}")
        print("Run ID:", RUN_ID)
        # Track Best Model
        if r2_test > best_r2:
            best_r2 = r2_test
            best_run_id = mlflow.active_run().info.run_id
            best_model_name = name
            best_model_uri = model_info.model_uri

# --- MODEL REGISTRATION ---
print(f"\nðŸ† Best Model: {best_model_name} (R2: {best_r2:.4f})")
print(f"ðŸ”— Best Run ID: {best_run_id}")
print(f"ðŸ”— Model Artifact URI: {best_model_uri}")

# Register using the exact URI returned by log_model
# This is safer than constructing the string manually
registered_model = mlflow.register_model(best_model_uri, "BikeRentalRegressor_2011")

# Transition to Staging
client = mlflow.tracking.MlflowClient()
client.transition_model_version_stage(
    name="BikeRentalRegressor_2011",
    version=registered_model.version,
    stage="Staging"
)

print(f"\nâœ… Model Registered Successfully!")
print(f"   - Name: BikeRentalRegressor_2011")
print(f"   - Version: {registered_model.version}")
print(f"   - Stage: Staging")
print(f"   - Load URI for Task 2: 'models:/BikeRentalRegressor_2011/Staging'")
print(f"   - View report at: http://localhost:5000 (Artifacts tab)")
```

```python
# --- Calculate Performance Metrics for Decision Matrix ---
# Baseline from Task 1
r2_2011_baseline = 0.7084
rmse_2011_baseline = 784.10
mae_2011_baseline = 608.68
# Current performance on 2012
r2_2012 = r2_score(y_2012, y_pred_2012)
rmse_2012 = np.sqrt(mean_squared_error(y_2012, y_pred_2012))

# Calculate changes
r2_drop_pct = ((r2_2011_baseline - r2_2012) / r2_2011_baseline) * 100
rmse_increase_pct = ((rmse_2012 - rmse_2011_baseline) / rmse_2011_baseline) * 100

print(f"ðŸ“Š Performance Metrics Calculated:")
print(f"   RÂ² Drop: {r2_drop_pct:.1f}%")
print(f"   RMSE Increase: {rmse_increase_pct:.1f}%")
```

```python
# Decision thresholds
THRESHOLD_R2_DROP = 10.0
THRESHOLD_RMSE_INCREASE = 15.0

signals = 0
if r2_drop_pct > THRESHOLD_R2_DROP: signals += 1
if rmse_increase_pct > THRESHOLD_RMSE_INCREASE: signals += 1

if signals >= 2:
    print("ðŸš¨ RECOMMENDATION: RETRAIN IMMEDIATELY")
else:
    print("âš ï¸ RECOMMENDATION: MONITOR CLOSELY")
```

```python
print("\n" + "="*80)
print("ðŸ“Š PERFORMANCE DEGRADATION ANALYSIS")
print("="*80)

# Current performance on 2012 (you already calculated y_pred_2012)
r2_2012 = r2_score(y_2012, y_pred_2012)
rmse_2012 = np.sqrt(mean_squared_error(y_2012, y_pred_2012))
mae_2012 = mean_absolute_error(y_2012, y_pred_2012)

# Calculate degradation
r2_drop = r2_2011_baseline - r2_2012
r2_drop_pct = (r2_drop / r2_2011_baseline) * 100
rmse_increase = rmse_2012 - rmse_2011_baseline
rmse_increase_pct = (rmse_increase / rmse_2011_baseline) * 100
mae_increase = mae_2012 - mae_2011_baseline
mae_increase_pct = (mae_increase / mae_2011_baseline) * 100

# Display comparison table
comparison_df = pd.DataFrame({
    'Metric': ['RÂ²', 'RMSE', 'MAE'],
    '2011 (Baseline)': [f"{r2_2011_baseline:.4f}", 
                        f"{rmse_2011_baseline:.2f}", 
                        f"{mae_2011_baseline:.2f}"],
    '2012 (Production)': [f"{r2_2012:.4f}", 
                          f"{rmse_2012:.2f}", 
                          f"{mae_2012:.2f}"],
    'Change': [f"{r2_drop:+.4f} ({r2_drop_pct:+.1f}%)",
               f"{rmse_increase:+.2f} ({rmse_increase_pct:+.1f}%)",
               f"{mae_increase_pct:+.1f}%"]
})

print("\n" + comparison_df.to_string(index=False))

# Interpretation
print(f"\nðŸ’¡ Interpretation:")
if r2_drop_pct > 10:
    print(f"   ðŸ”´ SEVERE degradation: RÂ² dropped by {r2_drop_pct:.1f}%")
    print(f"      Model explains {r2_drop_pct:.1f}% LESS variance in 2012")
elif r2_drop_pct > 5:
    print(f"   ðŸŸ¡ MODERATE degradation: RÂ² dropped by {r2_drop_pct:.1f}%")
else:
    print(f"   ðŸŸ¢ MINOR degradation: RÂ² dropped by {r2_drop_pct:.1f}%")

if rmse_increase_pct > 15:
    print(f"   ðŸ”´ CRITICAL: RMSE increased by {rmse_increase_pct:.1f}%")
    print(f"      Daily planning errors increased by {rmse_increase:.0f} bikes/day")
elif rmse_increase_pct > 10:
    print(f"   ðŸŸ¡ SIGNIFICANT: RMSE increased by {rmse_increase_pct:.1f}%")
else:
    print(f"   ðŸŸ¢ ACCEPTABLE: RMSE increased by {rmse_increase_pct:.1f}%")
```

```python
print("\n" + "="*80)
print("ðŸ”— CONNECTING DRIFT TO PREDICTION ERRORS")
print("="*80)

# Evidently identified these features as drifted
drifted_features = ['weathersit', 'hum_transformed']

# Calculate residuals
residuals_2012 = y_2012 - y_pred_2012
abs_residuals = np.abs(residuals_2012)

if 'residuals' not in df_2012_clean.columns:
    df_2012_clean['residuals'] = residuals_2012

# Analyze each drifted feature's impact
print("\nAnalyzing Drifted Features:")
print("-" * 60)

for feature in drifted_features:
    if feature == 'hum_transformed':
        feature_values = df_2012_clean['hum_transformed'].values
    elif feature == 'weathersit':
        feature_values = df_2012_clean['weathersit'].values
    else:
        continue
    
    # Calculate correlation between feature and errors
    corr = np.corrcoef(feature_values, abs_residuals)[0, 1]
    
    print(f"\n{feature}:")
    print(f"   â€¢ Correlation with errors: {corr:+.3f}")
    
    if abs(corr) > 0.3:
        print(f"   â€¢ ðŸ”´ STRONG impact on errors")
        if corr > 0:
            print(f"      â†’ Higher {feature} values â†’ Larger errors")
        else:
            print(f"      â†’ Lower {feature} values â†’ Larger errors")
    elif abs(corr) > 0.1:
        print(f"   â€¢ ðŸŸ¡ MODERATE impact on errors")
    else:
        print(f"   â€¢ ðŸŸ¢ WEAK impact on errors")

# Compare error distribution across weather conditions
print("\nðŸ“Š Error Breakdown by Weather Situation:")

weather_errors = df_2012_clean.groupby('weathersit').agg({
    'cnt': 'count'
}).rename(columns={'cnt': 'Days'})

weather_errors['Mean_Error'] = df_2012_clean.groupby('weathersit')['residuals'].apply(
    lambda x: np.abs(x).mean()
)

weather_labels = {
    1: 'Clear/Partly Cloudy',
    2: 'Mist/Cloudy',
    3: 'Light Rain/Snow'
}
weather_errors['Weather'] = weather_errors.index.map(weather_labels)

print(weather_errors[['Weather', 'Days', 'Mean_Error']].to_string(index=False))

worst_weather = weather_errors['Mean_Error'].idxmax()
print(f"\nðŸŒ§ï¸ Model struggles MOST in: {weather_labels[worst_weather]}")
print(f"   Mean error: {weather_errors.loc[worst_weather, 'Mean_Error']:.0f} bikes")
```

```python
import mlflow
import mlflow.sklearn
from mlflow.models.signature import infer_signature

mlflow.set_experiment("Task2_Drift_Retraining")

with mlflow.start_run(run_name="rf_retrained") as run:
    
    # 1. LOGGING PARAMETERS
    mlflow.log_params(rf_model.get_params())
    mlflow.log_param("retraining rf model", "2012_Dataset")
    
    # 2. LOG METRICS
    rmse_test = np.sqrt(mean_squared_error(y_test, rf_pred))
    r2_test = r2_score(y_test, rf_pred)
    mae_test = mean_absolute_error(y_test, rf_pred)
    rmse_train = np.sqrt(mean_squared_error(y_train, rf_pred_train))
    r2_train = r2_score(y_train, rf_pred_train)
    mae_train = mean_absolute_error(y_train, rf_pred_train)
    
    mlflow.log_metric("test_rmse", rmse_test)
    mlflow.log_metric("test_r2", r2_test)
    mlflow.log_metric("test_mae", mae_test)
    mlflow.log_metric("train_rmse", rmse_train)
    mlflow.log_metric("train_r2", r2_train)
    mlflow.log_metric("train_mae", mae_train)
    
    print(f"Metrics Logged -> RMSE: {rmse_test:.2f}, R2: {r2_test:.4f}, MAE: {mae_test:.2f}")

    # 3.LOG ARTIFACTS (Visualizations)
    
    mlflow.log_artifact("residual_plot_retrained.png")
    mlflow.log_artifact("rf_model_feature_importance_retrained.png")
    mlflow.log_artifact("feature_impact_retrained.png")
    mlflow.log_artifact("shap_summary_plot_retrained.png")
    mlflow.log_artifact("day_of_year_retrained.png")
    mlflow.log_artifact("temperature_plot_retrained.png")
    mlflow.log_artifact("unscale_temperature_plot_retrained.png")

    joblib.dump(scaler, "scaler_2012.joblib")
    mlflow.log_artifact("scaler_2012.joblib")
    
    # 4. INFER SIGNATURE & REGISTER MODEL
    # This defines the input/output schema for the model
    signature = infer_signature(X_test, rf_pred)
    
    model_info = mlflow.sklearn.log_model(
        sk_model=rf_model,
        artifact_path="model",
        signature=signature,
        registered_model_name="BikeRental_Retrained_2012_RF"  
    )
    
    print(f"\nâœ… Run Complete. Model Registered as 'BikeRental_Retrained_2012_RF'.")
    print(f"Run ID: {run.info.run_id}")
    print(f"   - View report at: http://localhost:5000 (Artifacts tab)")
```
